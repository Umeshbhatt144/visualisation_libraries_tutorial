```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = sns.load_dataset("flights")
sns.relplot(x="passengers", y="month",data=a)
```

```python
a = sns.load_dataset("flights")
sns.relplot(x="passengers", y="month",hue="year",data=a)
```

```python
b = sns.load_dataset("tips")
sns.relplot(x="time", y="tip", data=b, kind="line")
```

```python
sns.catplot(x="day", y="total_bill", data=b)
```

```python
sns.catplot(x="day", y="total_bill", kind="violin", data=b)
```

```python
sns.catplot(x="day", y="total_bill", kind="boxen", data=b)
```

```python
x = pd.DataFrame({'Day': [1,2,3,4,5,6,7], 'Grocery':[30,20,45,80,51,46,78], 'Clothes':[13,40,54,63,74,66,45], 'Utensils':[12,32,14,65,75,45,64]})
y = pd.DataFrame({'Day': [8,9,10,11,12,13,14], 'Grocery':[30,20,45,80,51,46,78], 'Clothes':[13,40,54,63,74,66,45], 'Utensils':[12,32,14,65,75,45,64]})
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data= np.random.multivariate_normal(mean, cov, 200)
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind = "kde", color ="b");
```

Multi_Plot grid

```python
a = sns.load_dataset("iris")
b = sns.FacetGrid(a, col="species")
b.map(plt.hist, "sepal_length")
```

```python
a = sns.load_dataset("flights")
b = sns.PairGrid(a)
b.map(plt.scatter)
```

Plot-Aesthetics

```python
sns.set(style="darkgrid")
a = sns.load_dataset("flights")
b = sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style="dark")
a = sns.load_dataset("flights")
b = sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style="white", color_codes=True)
a=sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=a)
```

```python
sns.set(style="white", color_codes=True)
a=sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=a)
sns.despine(offset=10, trim=True)
```
