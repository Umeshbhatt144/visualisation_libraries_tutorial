Matplotlib is a python library used for data visualization

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

Line Plot

```python
x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

```python
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.plot(x,y,color='y',linestyle=':',linewidth=2)
plt.show()
```

Adding two items in the same list

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x

plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.stackplot(x, y1, color='purple', alpha=0.8)

plt.plot(x,y2,color='r',linestyle='-',linewidth=3)
plt.stackplot(x, y2, color='blue', alpha=0.2)

plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
# plt.grid(True)
plt.show()
```

Adding Subplots

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
plt.subplot(1,2,1)              # subplot(rows,column,index)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=3)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
plt.subplot(2,1,1)              # subplot(rows,column,index)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(2,1,2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=3)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

# Bar Plot

```python
student = {'Bob':45, 'Matt':56, "Sam":73}
names = list(student.keys())
values = list(student.values())
plt.bar(names,values)
plt.title('Bar Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
student = {'Bob':45, 'Matt':56, "Sam":73}
names = list(student.keys())
values = list(student.values())
plt.barh(names,values)       # add h in last to convert into horizontal bar graph
plt.title('Bar Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

# Scatter Plot

```python
x= [4,2,5,1,3,6,9,7]
y= [1,6,2,7,8,3,4,9]
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x= [4,2,5,1,3,6,9,7]
y= [1,6,2,7,8,3,4,9]
plt.scatter(x,y,marker='*',c='g',s=100)
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x= [10,20,30,40,50,60,70,80]
a= [1,6,2,7,8,3,4,9]
b= [5,1,4,6,9,3,2,7]
plt.scatter(x,a,marker='.',c='r',s=200)
plt.scatter(x,b,marker='4',c='g',s=100)
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x= [10,20,30,40,50,60,70,80]
a= [1,6,2,7,8,3,4,9]
b= [5,1,4,6,9,3,2,7]
plt.subplot(1,2,1)
plt.scatter(x,a,marker='.',c='r',s=200)
plt.subplot(1,2,2)
plt.scatter(x,b,marker='*',c='g',s=100)
plt.show()
```

Histogram

```python
data = [1,3,1,3,4,6,5,8,3,4,1,5,8,9]
plt.hist(data)
plt.show()
```

```python
data = [1,3,1,3,4,6,5,8,3,4,1,5,8,9]
plt.hist(data, color='g', bins=3)
plt.show()
```

Box-Plot

```python
one= [1,2,3,4,5,6,7,8,9]
two= [1,2,3,4,5,4,3,2,1]
three= [6,7,8,9,8,7,6,5,4]
data= list([one,two,three])
plt.boxplot(data)
plt.show()
```

Violin-Plot

```python
one= [1,2,3,4,5,6,7,8,9]
two= [1,2,3,4,5,4,3,2,1]
three= [6,7,8,9,8,7,6,5,4]
data= list([one,two,three])
plt.violinplot(data)
plt.show()
```

```python
one= [1,2,3,4,5,6,7,8,9]
two= [1,2,3,4,5,4,3,2,1]
three= [6,7,8,9,8,7,6,5,4]
data= list([one,two,three])
plt.violinplot(data,showmedians=True)
plt.show()
```

Pie-Chart

```python
fruit= ['Apple', 'Orange', 'Mango', 'Guava']
quantity = [67,89,56,45]
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['red','orange','yellow','green'], shadow=True,startangle=90, explode=(0.1,0.1,0,0))
plt.show()
```

Dough-nut Chart

```python
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([100],colors=['w'],radius=1)
plt.show()
```
