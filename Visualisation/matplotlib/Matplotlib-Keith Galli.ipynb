```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# keyword argument notation
plt.plot(x,y,label='2x',c='r',linewidth=2,linestyle='
         --', marker= '.',markersize=10, markeredgecolor='blue')

plt.title("Line Plot", fontdict = {'fontname': 'Comic Sans MS', 'fontsize':20})
# to change fontdesign and fontsizze of a title

plt.xlabel('X-axis',fontdict={'fontname': 'Arial'})
plt.ylabel('Y-axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.legend()
# plt.grid()
plt.show()
```

```python
# by using short hand notation
plt.plot(x,y, 'b^--', label='2x')
plt.title("Line Plot", fontdict = {'fontname': 'Comic Sans MS', 'fontsize':20})
# to change fontdesign and fontsizze of a title

plt.xlabel('X-axis',fontdict={'fontname': 'Arial'})
plt.ylabel('Y-axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.legend()
plt.show()
```

```python
plt.plot(x,y, 'b^--', label='2x')

x2 = np.arange(0,4.5,0.5)
plt.plot(x2, x2**2, 'r.-', label='x^2', linewidth=2, markersize=10)

plt.title("Line Plot", fontdict = {'fontname': 'Comic Sans MS', 'fontsize':20})
# to change fontdesign and fontsizze of a title

plt.xlabel('X-axis',fontdict={'fontname': 'Arial'})
plt.ylabel('Y-axis')
plt.xticks([0,1,2,3,4])


plt.legend()
plt.show()
```

```python
# Resize your graph
plt.figure(figsize=(5,3),dpi=100)
# dpi means pixel size

plt.plot(x,y, 'b^--', label='2x')

x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'r.-', label='x^2', linewidth=2, markersize=10)
plt.plot(x2[4:], x2[4:]**2, 'r.--',linewidth=2, markersize=10)

plt.title("Line Plot", fontdict = {'fontname': 'Comic Sans MS', 'fontsize':20})
# to change fontdesign and fontsizze of a title

plt.xlabel('X-axis',fontdict={'fontname': 'Arial'})
plt.ylabel('Y-axis')
plt.xticks([0,1,2,3,4])
plt.legend()


# to save figure in your system
plt.savefig('mygraph.png', dpi=300)

plt.show()
```

Bar chart

```python
labels = ["A", "B", "C"]
values = [1,4,2]

bar = plt.bar(labels, values)

bar[0].set_hatch('/')
# bar[1].set_hatch('--')
bar[1].set_hatch('*')
bar[2].set_hatch('O')
plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ["A", "B", "C"]
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ["/", "O", "*"]
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    
plt.figure(figsize=(6,4))

plt.show()
```

```python
t1 = np.arange(0.0, 2.0, 0.1)
t2 = np.arange(0.0, 2.0, 0.01)

# note that plot returns a list of lines.  The "l1, = plot" usage
# extracts the first element of the list into l1 using tuple
# unpacking.  So l1 is a Line2D instance, not a sequence of lines
l1, = plt.plot(t2, np.exp(-t2))
l2, l3 = plt.plot(t2, np.sin(2 * np.pi * t2), '--go', t1, np.log(1 + t1), '.')
l4, = plt.plot(t2, np.exp(-t2) * np.sin(2 * np.pi * t2), 'rs-.')

plt.legend((l2, l4), ('oscillatory', 'damped'), loc='upper right', shadow=True)
plt.xlabel('time')
plt.ylabel('volts')
plt.title('Damped oscillation')
plt.show()
```
